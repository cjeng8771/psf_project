```python
from defocus import Microscope
import matplotlib.pyplot as plt
import torch
```

```python
device = torch.device('cpu') # change this to gpu as needed
```

```python
scope = Microscope(dim=512, NA=0.5, mag=30, wavelength=500e-9, pixel_size=2e-6, device=device)
```

```python
psf_0 = scope.get_psf(defocus_length=0) # in focus
psf_1 = scope.get_psf(defocus_length=5e-6) # 5 microns out of focus
psf_2 = scope.get_psf(defocus_length=10e-6) # 10 microns out of focus


crop=100 #cropping to zoom in for visualization purposes
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(psf_0.cpu()[crop:-crop,crop:-crop], cmap='magma')
plt.subplot(1,3,2)
plt.imshow(psf_1.cpu()[crop:-crop,crop:-crop], cmap='magma')
plt.subplot(1,3,3)
plt.imshow(psf_2.cpu()[crop:-crop,crop:-crop], cmap='magma')
```
